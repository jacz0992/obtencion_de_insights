### Análisis exploratorio EDA

```python
data.columns
```

```python
#Necesitamos saber cuales son los nombres unicos de los bancos en la lista
nombresBancos = data['NOMBRE DE LA  ENTIDAD'].unique().tolist()
nombresBancos
```

```python
### cuantos usuarios hay por Banco

fig, ax = plt.subplots(1,1, figsize=(20,10))
plt.yticks(range(0, 100000, 2000))
ax = data['NOMBRE DE LA  ENTIDAD'].value_counts().plot(kind="bar")
ax.set_ylabel("Cantidad de clientes ")
ax.set_xlabel("Nombre de la Entidad")
for p in ax.patches:
    ax.annotate(
        str(p.get_height()), xy=(p.get_x() + 0.25, p.get_height() + 0.1), fontsize=12
    )
```

```python
#me interesa saber del banco Banagrario que tipo de cuentas tiene

banagra = data.loc[data['NOMBRE DE LA  ENTIDAD'] == 'Banagrario']
fig, ax = plt.subplots(1,1, figsize=(20,10))
plt.yticks(range(0, 15000, 1000))
banagra['TIPO'].value_counts().plot(kind="bar")
ax.set_ylabel("Cantidad de cuentas ")
ax.set_xlabel("Tipo de cuenta del Banagrario")
for p in ax.patches:
    ax.annotate(
        str(p.get_height()), xy=(p.get_x() + 0.25, p.get_height() + 0.2), fontsize=12
    )
```

```python
### Ahora teniendo la lista de cuentas, necesito saber el tipo de corresponsales en donde se encuentran ubicadas


data.rename(columns={'NRO CORRESPONSALES PROPIOS':'NRO_CORRESPONSALES_PROPIOS'},
               inplace=True)
data.columns

tableCross1 = pd.crosstab(banagra['TIPO'], banagra['DEPARTAMENTO'])
tableCross1

```

```python
#CREARE UN HISTOGRAMA QUE PERMITA MEDIR LA DENSIDAD POR DEPARTAMENTO
filMicrocreditos = tableCross1.loc['MICROCREDITO']
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,5))

sns.lineplot(data=filMicrocreditos)
ax.tick_params(axis='x', rotation=60)

```
